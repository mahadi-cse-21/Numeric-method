```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return x*x - 4*x - 10

# Initialize x values for plotting
x = np.arange(-6.0, 6.0, 0.25)
y = []

# Calculate the corresponding y values for each x
for i in x:
    y.append(f(i))

# Bisection method to find the root
low = -6
high = 6
tolerance = 0.00001  # Stopping condition tolerance

# Perform the bisection method
while abs(high - low) > tolerance:
    x0 = (low + high) / 2
    if f(x0) * f(low) < 0:
        high = x0
    else:
        low = x0

# Print the root found by the bisection method
print("Root found:", (low + high) / 2)

# Plot the function
plt.plot(x, y)
plt.grid(True)
plt.axhline(y=0, color="black")  # Add horizontal line at y = 0
plt.axvline(x=0, color="black")  # Add vertical line at x = 0
plt.xlabel("x")
plt.ylabel("y = f(x)")
plt.title("Plot of y = x^2 - 4x - 10")
plt.show()
```
