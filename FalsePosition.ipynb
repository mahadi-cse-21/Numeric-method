```python
def f(x):
    return x*x - 4*x - 10

x1 = float(input("Give first guess: "))
x2 = float(input("Give second guess: "))

tolerance = 0.00001

while abs(x2 - x1) > tolerance:
    x0 = x1 - (f(x1) * (x2 - x1)) / (f(x2) - f(x1))
    print(x1)
    print(x2)
    print(f(x1))
    print(f(x2))
    print(x0)
    print(f(x0))
    if f(x0) * f(x1) < 0:
        x2 = x0
    else:
        x1 = x0

print(f"The root is approximately: {x0}")
```
