```python
def f(x):
    return 4*pow(x, 4) + 3*x*x + 2*x + 5

def trapezoidal(x1, x2, h):
    n = int((x2 - x1) / h)
    x = []
    y = []
    i = x1
    while i <= x2:
        x.append(i)
        i += h
    for i in range(len(x)):
        y.append(f(x[i]))

    sum = 0.0
    for i in range(1, len(x) - 1):
        sum += 2 * y[i]

    sum += y[0] + y[n]
    
    return (h * sum) / 2

def simpson_1_3(x1, x2, h):
    n = int((x2 - x1) / h)
    if n % 2 != 0:
        n += 1
    x = []
    y = []
    i = x1
    while i <= x2:
        x.append(i)
        i += h
    for i in range(len(x)):
        y.append(f(x[i]))

    sum = y[0] + y[n]

    for i in range(1, len(x) - 1, 2):
        sum += 4 * y[i]
    for i in range(2, len(x) - 1, 2):
        sum += 2 * y[i]
    
    return (h / 3) * sum
def simpson_3_8(x1, x2, h):
    n = int((x2 - x1) / h)
    if n % 2 != 0:
        n += 1
    x = []
    y = []
    i = x1
    while i <= x2:
        x.append(i)
        i += h
    for i in range(len(x)):
        y.append(f(x[i]))

    sum = y[0] + y[n] 
    
    for i in range(3, len(x) - 1, 3):
        sum += 2 * y[i]
    for i in range(1, len(x) - 1, 1):
        if i%3==0:
            continue
        sum += 3 * y[i]
    
    return (3 * h / 8) * sum
    
    



x1 = 2
x2 = 6
h = 0.25

print(f"After applying trapezoidal rule: {trapezoidal(x1, x2, h)}\n")
print(f"After applying Simpson's 1/3 rule: {simpson_1_3(x1, x2, h)}\n")
print(f"After applying Simpson's 3/8 rule: {simpson_3_8(x1, x2, h)}\n")
```
