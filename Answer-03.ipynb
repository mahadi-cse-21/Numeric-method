```python
x= []
y=[]
n=5
for i in range(n):
    a = float(input("Input the  value of  x"))
    x.append(a)
    b = float(input("Input corresponding value of y according to x"))
    y.append(b)
print(x)
print(y)
```

```python

sumOfxiyi = 0.0
sumofxi=0.0
sumofyi=0.0
sumofxi2=0.0
for i in range(n):
    sumofxi+=x[i]
    sumofyi+=y[i]
    sumOfxiyi+=x[i]*y[i]
    sumofxi2+=x[i]*x[i]

m = (n*sumOfxiyi - (sumofxi*sumofyi))/(n*sumofxi2-(sumofxi*sumofxi))

b = (sumofyi - m*sumofxi)/n

print(f"Regression Line Equation is : y = {m} x + {b}")

import matplotlib.pyplot as plt

plt.plot(x,y,'o',color="red")
plt.grid()

plt.axhline(y=0,color="black")
plt.axvline(x=0,color="black")
plt.title(f"Y = {m} * X +  {b} Regression Line")
plt.xlabel("X = Number of rooms of a house")
plt.ylabel("Y = Rent of a house according number of  room ")
plt.show()
```
